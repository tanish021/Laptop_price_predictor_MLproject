```python
# Clean memory column
df['Memory'] = df['Memory'].astype(str).replace('\.0', '', regex=True)
df["Memory"] = df["Memory"].str.replace('GB', '', regex=False)
df["Memory"] = df["Memory"].str.replace('TB', '000', regex=False)

# Split memory layers
new = df["Memory"].str.split(r"\+", n=1, expand=True)

df["first"] = new[0].str.strip()
df["second"] = new[1]

# Identify storage types in first layer
df["Layer1HDD"] = df["first"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer1SSD"] = df["first"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer1Hybrid"] = df["first"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer1Flash_Storage"] = df["first"].apply(lambda x: 1 if "Flash Storage" in x else 0)

# Extract numeric size
df['first'] = df['first'].str.replace(r'\D', '', regex=True)

# Handle second layer
df["second"].fillna("0", inplace=True)

df["Layer2HDD"] = df["second"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer2SSD"] = df["second"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer2Hybrid"] = df["second"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer2Flash_Storage"] = df["second"].apply(lambda x: 1 if "Flash Storage" in x else 0)

df['second'] = df['second'].str.replace(r'\D', '', regex=True)

# Convert to integers
df["first"] = df["first"].astype(int)
df["second"] = df["second"].astype(int)

# Calculate final storage values
df["HDD"] = (df["first"] * df["Layer1HDD"] + df["second"] * df["Layer2HDD"])
df["SSD"] = (df["first"] * df["Layer1SSD"] + df["second"] * df["Layer2SSD"])
df["Hybrid"] = (df["first"] * df["Layer1Hybrid"] + df["second"] * df["Layer2Hybrid"])
df["Flash_Storage"] = (df["first"] * df["Layer1Flash_Storage"] + df["second"] * df["Layer2Flash_Storage"])

# Drop temporary columns
df.drop(columns=[
    'first', 'second',
    'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid', 'Layer1Flash_Storage',
    'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid', 'Layer2Flash_Storage'
], inplace=True)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error

# Automatically detect all categorical (text) columns from the training data
# select_dtypes(include='object') selects columns containing strings like "Windows", "Dell"
categorical_cols = x_train.select_dtypes(include='object').columns


# Create a ColumnTransformer to preprocess the dataset
step1 = ColumnTransformer(
    transformers=[
        # 'cat' is the name of this transformation step
        # OneHotEncoder converts categorical text columns into numerical columns
        # drop='first' removes the first category to avoid dummy variable trap
        # sparse_output=False ensures output is a normal array instead of sparse matrix
        # categorical_cols tells which columns need encoding
        ('cat', OneHotEncoder(drop='first', sparse_output=False), categorical_cols)
    ],

    # remainder='passthrough' keeps all other numerical columns unchanged
    remainder='passthrough'
)


# Create the Linear Regression model which will learn the relationship
# between input features (RAM, Company, OS etc.) and the target (Price)
step2 = LinearRegression()


# Create a Pipeline to combine preprocessing and model training
pipe = Pipeline([

    # First step: apply ColumnTransformer to encode categorical data
    ('step1', step1),

    # Second step: train the Linear Regression model
    ('step2', step2)

])


# Train the pipeline using training data
# Internally two things happen:
# 1) Categorical columns are encoded
# 2) Linear Regression model learns patterns from the data
pipe.fit(x_train, y_train)


# Use the trained pipeline to make predictions on the test dataset
# The same preprocessing (encoding) is automatically applied to test data
y_pred = pipe.predict(x_test)


# Print the R2 score which measures how well the model explains the data
# Higher R2 means better model performance
print("R2 score:", r2_score(y_test, y_pred))


# Print the Mean Absolute Error (MAE)
# It shows the average difference between actual price and predicted price
# Lower MAE means more accurate predictions
print("MAE:", mean_absolute_error(y_test, y_pred))

```
